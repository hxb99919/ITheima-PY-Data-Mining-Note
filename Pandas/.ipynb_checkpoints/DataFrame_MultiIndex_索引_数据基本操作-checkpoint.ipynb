# DataFrame

```python
import numpy as np
import pandas as pd

# 创建一个符合正态分布的10个股票5天的涨跌幅度
stock_change = np.random.normal(0, 1, (10, 5))
```

```python
stock_change
```

```python
# 使用Pandas中的数据结构
pd.DataFrame(stock_change)
```

```python
# 添加行索引
stock = ["股票{}".format(i) for i in range(10)] 
```

```python
pd.DataFrame(stock_change, index=stock)
```

```python
# 添加列索引
date = pd.date_range(start="20180101", periods=5, freq="B")
```

```python
date
```

```python
data = pd.DataFrame(stock_change, index=stock, columns=date)
```

```python
data
```

## 修改行列索引值

```python
data.head()    # 不给参数，默认给5组
```

```python
'''
如果输入 
data.index[2] = "股票88" 
会报错，因为无法单独对一个index进行修改
''' 
```

```python
stock_ = ["股票_{}".format(i) for i in range(10)] 
```

```python
data
```

```python
""" 
重设索引
reset_index(drop=False)
* 设置新的下标索引
* drop： 默认为False，不删除原来索引，如果为True，则删除
"""
```

```python
data.reset_index(drop=False)    # 新的索引在旧的索引左边
```

```python
data.reset_index(drop=True)    # 旧的索引被删除
```

## 设置新索引

```python
df = pd.DataFrame({'month': [1, 4, 7, 10],
                   'year' : [2012, 2014, 2013, 2014],
                  'sale' : [55, 40, 84, 31] })
```

```python
df
```

```python
# 以月份设置新的索引
# set_index(keys, drop=True)
# keys:列索引名称或列索引名称的列表
# drop:默认为True，当做新的索引，删除旧的索引
df.set_index("month", drop=False)
```

```python
df.set_index("month", drop=True)
```

```python
new_df = df.set_index(["month", "year" ])
```

```python
new_df.index    # 行名称
```

```python
new_df.index.names
```

```python
new_df.index.levels
```

```python
# data["2018-02-26"]["open"] 会报错，必须先列后行
```

```python
data.loc["2018-02-26"]["open"]    # loc按名字索引
```

```python
data.loc["2018-02-26", "open"]
```

```python
data.iloc[1, 0]    # iloc按数字索引
```

```python
# 获取行第一天到第四天，["open", "close", "high", "low"]这4个指标的结果
# data.ix[:4, ["open", "close", "high", "low"]] 
# 在pandas的1.0.0版本开始，移除了Series.ix and DataFrame.ix 方法。
# 我的解决方法：使用DataFrame的loc方法或者iloc方法进行替换吧

data.loc[data.index[0:4], ['open', 'close', 'high', 'low']]
data.iloc[0:4, data.columns.get_indexer(['open', 'close', 'high', 'low'])]
```

## 逻辑运算符号

```python
# 例如筛选p_change > 2的日期数据
data[data["p_change"] > 2].head()
```

```python
# 完成一个多逻辑判断，筛选p_change > 2并且low > 15
data[(data["p_change"]) > 2  & (data["low"]>15)].head()
```

```python
data.query("p_change > 2 & low > 15" )
```

```python
# 判断'turnover'是否为4.19，2.39
data[data["turnover"].isin([4.19, 2.39])]
```
