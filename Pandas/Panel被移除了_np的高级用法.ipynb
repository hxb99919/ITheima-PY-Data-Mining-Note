## 数据离散化

```python
# 1准备数据
data = pd.Series([165, 174, 160, 180, 159, 163, 192 ,184], index=['No1:165','No2:174','No3:160','No4:180','No5:159','No6:163','No7:192','No8:184',])
```

```python
data
```

```python
# 2分组
# 自动分组
sr =pd.qcut(data, 3)
```

```python
type(sr)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# 3转换成one-hot编码
pd.get_dummies(sr, prefix="height")
```

```python
# 自定义分组
bins = [150, 165, 180, 192]
sr = pd.cut(data, bins)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# get_dummies
pd.get_dummies(sr, prefix="身高")
```
