### 柱状图对比每部电影的票房收入

```python
import matplotlib.pyplot as plt
# 1、准备数据
movie_names = ['雷神3：诸神黄昏', '正义联盟', '东方快车谋杀案', '寻梦环游记', 
               '全球风暴', '降魔传', '追捕', '七十七天', '密战', '狂兽', '其他']
tickets = [73857, 57767, 22354, 15696, 14839, 8725, 8716, 8318, 7916, 6747, 52222]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制柱状图
x_ticks = range(len(movie_names))
plt.bar(x_ticks, tickets, color=['b','r','g','y','c','m','y','k','c','g','b'])

# 修改x刻度
plt.xticks(x_ticks, movie_names)

# 添加标题
plt.title("电影票房收入对比")

# 添加网格显示
plt.grid(linestyle='--', alpha=0.5)

# 4、显示图像
plt.show()
```

### 直方图绘制

```python
# 直方图最重要的就是组数和组距
# 1、准备数据
movie_number = range(250)
movie_time = [int(random.uniform(75, 150)) for i in movie_number]   # 电影时长
# 数据这里因为需要250个，所以就随机生成了，然后电影时长是参考视频里的大致范围，也符合常识，但每次运行数据不一样
# 如果random没有转int，会在修改轴刻度时会报错 —— float 不能转换成int

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制直方图
distance = 2    # 组距
group_number = int((max(movie_time)-min(movie_time))/distance)       

# matplotlib.pylot.hist(x, bin=None, normed=None, **kargs)    
plt.hist(movie_time, bins=group_number)    # bins=组数
# 'normed=True'代表频率值即百分比，但已经过时，改成'density=True'

# 修改x轴刻度
plt.xticks(range(min(movie_time), max(movie_time) + 2, distance))    # 这里+2是为了让直方图的最右端完全显示在x轴数字内部    

# 添加标题
plt.xlabel("电影时长大小")
plt.ylabel("电影的数量")
plt.title("电影时长分布")

# 添加网格显示
plt.grid(linestyle='--', alpha=0.5)

# 4、显示图像
plt.show()
```

### 饼图

```python
# 1、准备数据
movie_name = ['雷神3：诸神黄昏', '正义联盟', '东方快车谋杀案', '寻梦环游记', 
              '全球风暴', '降魔传', '追捕', '七十七天', '密战', '狂兽', '其他']

place_count = [60605, 54546, 45819, 28243, 13270, 
               9945, 7679, 6799, 6101, 4621, 20105]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制柱状图
# plt.pie(x, labels=, autopct=, colors)
'''
x:数量，自动计算百分比  labels：每部分名称
autopct:占比显示指定%.2f%%  保留两位小数，附带上百分号  colors:每部分颜色
'''
plt.pie(place_count, labels=movie_name, autopct="%.2f%%", colors=['b','r','g','y','c','m','y','k','c','g','b'])

plt.axis('equal')    #让饼图保持正圆

# 添加标题
plt.title("电影排片占比")

# 显示图例
plt.legend()

# 4、显示图像
plt.show()
```
